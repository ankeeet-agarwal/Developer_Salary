### 3.8 Explanatory Model (Multiple Linear Regression)

Based on our findings in the exploratory analysis, we shall consider the following variables for predictive analytics :-
    

    1) Years of Coding Experience
    2) Number of Languages known
    3) Number of databases known
    4) Number of platforms known
    5) Number of webframes known
    6) Number of Technology tools used
    7) Education attained
    8) Annual Salary (dependent variable)
    
    Based on this the following regression equation is used for the explanatory model
    

$Annual Salary = \beta_0 + \beta_1*YearsCodePro + \beta_3*LangCount + \beta_4*databaseCount + \beta_5*PlatformCount + \beta_6*WebframeCount + \beta_7*TechtoolsCount + \beta_8*NewToolCount + \beta_9*4yeardegree + \beta_10*Advanceddegree + \beta_11*Somecollege + \epsilon$

```python

#Create Dataframes for X and Y
y = dfcorr['ConvertedCompYearly']
X = pd.concat([dfcorr, pd.get_dummies(dfcorr['Edu'])], axis=1)
X['Const'] = 1
X = X.drop(['HighSchool_or_less', 'Edu', 'ConvertedCompYearly', 'EdLevel', 'Age' ], axis=1)
X
```

```python
# Fit a full model
mod_full = sm.OLS(y,X).fit()

# Summarize model
mod_full.summary()
```

#### Improving the Regression Model by removing the insignificant predictors

The dummy-coded variables for education are retained as they each explain the effect of levels of education attainment

```python
# Fit refined models by dropping insignificant predictors
mod_refined1 = sm.OLS(y,X.drop(['database_Count' ],axis=1)).fit()

# Summarize model
mod_refined1.summary()
```

#### Multicollinearity Check

We find that education dummy-coded variables have high Variance Inflation Factor, therefore it is not recommended to explain developer's salary with these variables. Another Model is built to without education variables and the 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Further drop KM2 from the data
X2 = X.drop(['database_Count'],axis=1)
# Calculate VIF
pd.Series([variance_inflation_factor(X2.values, i)
           for i in range(X2.shape[1])], 
          index=X2.columns)
```

```python
#Remove
X3 = X2.drop(['4year_degree', 'Some_College'],axis=1)

pd.Series([variance_inflation_factor(X3.values, i)
           for i in range(X3.shape[1])], 
          index=X3.columns)
```

```python
mod_refined3 = sm.OLS(y,X3).fit()

mod_refined3.summary()
```

#### Model Comparison

This comparison shows that Model 3 is the most reliable for explaining developer's salary as it has similar adj R-squared value and has no multicollinearity issue. 

```python
# Use summary_col() to compare multiple models
from statsmodels.iolib.summary2 import summary_col
reg_sum = summary_col([mod_full,
                       mod_refined1,
                       mod_refined3],stars=True)
print(reg_sum)
```

### 4.2 K-Nearest Neighbors to predict salary

In this section, we first split the data into train and test sample. 60% of the rows are in training dataset and 40% are for testing. 

```python
#create X and Y dataframes
Knn_X = full_data.drop(['ConvertedCompYearly'], axis=1)
Knn_Y = full_data['ConvertedCompYearly']
```

```python
from sklearn.model_selection import train_test_split

# 60-40% simple split
# To make the result reproducible, set the random_state
train_y,test_y,train_X,test_X = train_test_split(Knn_Y, Knn_X,
                                                 test_size=0.4,
                                                 random_state=123)
```

```python
from sklearn import preprocessing

# Create a scaler to do the transformation
scaler = preprocessing.MinMaxScaler().fit(train_X)
# Transform training X
train_X_scale = scaler.transform(train_X)
train_X_scale = pd.DataFrame(train_X_scale)
train_X_scale.columns = train_X.columns

train_X_scale.describe().transpose()
```

```python
# Transform test X
test_X_scale = scaler.transform(test_X)
test_X_scale = pd.DataFrame(test_X_scale)
test_X_scale.columns = test_X.columns

test_X_scale.describe().transpose()
```

#### Hyperparameter Tuning

For K-nearest neighbors, we tune the hyperparameter between the ranges 25 to 45. This is because we observed that the value of Root Mean Square Error (RMSE) for values from k = 1 to k = 20 were declining. Here we find that, the RMSE estimate is lowest for k = 35. We therefore perform KNN Regression Analysis, using k = 35 and k = 36 to get a better understanding of the model performance estimates. We get better estimates in terms of R-Squared, Mean Absolute Percentage Error, and Root Mean Square Error.

```python
from sklearn import neighbors
from sklearn import metrics
import math
for k in range(25,45):
    k = k + 1
    knn = neighbors.KNeighborsRegressor(n_neighbors = k, 
                                         weights='uniform', 
                                         algorithm='auto')
    knn.fit(train_X_scale, train_y)
    pred_y = knn.predict(test_X_scale)
    print("RMSE is ", math.sqrt(metrics.mean_squared_error(test_y, pred_y))," for k =",k)
```

```python
# KNN: K=35, most optimum value as per tuning based on RMSE.
knn35 = neighbors.KNeighborsRegressor(n_neighbors=35)
knn35.fit(train_X_scale, train_y)
pred_y_knn35 = knn35.predict(test_X_scale)
```

```python
from sklearn.metrics import mean_absolute_percentage_error

#Mean absolute percentage error
print("MAPE is ",mean_absolute_percentage_error(test_y, pred_y_knn35), "for k =35")
```

```python
 print("RMSE is ", math.sqrt(metrics.mean_squared_error(test_y, pred_y_knn35))," for k =35")
```

```python
print("R-Squared is ", metrics.r2_score(test_y, pred_y_knn35)," for k =35")
```

```python
# KNN: K=36, most optimum value as per tuning based on RMSE.
knn36 = neighbors.KNeighborsRegressor(n_neighbors=36)
knn36.fit(train_X_scale, train_y)
pred_y_knn36 = knn36.predict(test_X_scale)
#Mean absolute percentage error
print("MAPE is ",mean_absolute_percentage_error(test_y, pred_y_knn36), " for k =36")
```

```python
print("RMSE is ", math.sqrt(metrics.mean_squared_error(test_y, pred_y_knn36))," for k =36")
```

```python
print("R-Squared is ", metrics.r2_score(test_y, pred_y_knn36)," for k =36")
```
